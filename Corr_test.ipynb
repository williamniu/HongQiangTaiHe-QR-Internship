```python
import pandas as pd
```

```python
# 读取数据
# s_benchmark = pd.read_feather('/home/user27/result/benchmark.fea')
s_benchmark = pd.read_feather('/home/user27/result/version226.fea')
s = pd.read_feather('/home/user27/result/version224.fea')  # 132 0.7710
s5 = pd.read_feather('/home/user27/result/version114.fea')
s7 = pd.read_feather('/home/user27/result/version118.fea')
s14 = pd.read_feather('/home/user27/result/version192.fea')
```

```python
# 设置索引
s_benchmark = s_benchmark.set_index('date')
s = s.set_index('date')

# 确保列对齐（重要！）
common_columns = s_benchmark.columns.intersection(s.columns)
s = s[common_columns]
s_benchmark = s_benchmark[common_columns]

# 转置数据框（行变列，列变行）
s_T = s.T
benchmark_T = s_benchmark.T

# 计算每行（原列）的相关系数
corr_series = s_T.corrwith(benchmark_T, method='pearson')
# 计算平均相关系数
average_corr = corr_series.mean()
```

```python
print(f"平均皮尔逊相关系数: {average_corr:.4f}")
print(f"详细相关系数统计:")
print(corr_series.describe())
```
